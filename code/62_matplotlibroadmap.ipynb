## Explicit plotting

## Chart types

```python
fig, axs = plt.subplots(
    1, 3,
    figsize=(16, 8)
)

data = [4, 2, 5, 1, 1, 4]

axs[0].set_title("Bar plot")
bars = axs[0].bar(
    x=range(len(data)),
    height=data,
    tick_label=list("ABCDEF"),
    color="magenta"
)


axs[1].set_title("Scatter plot")
points = axs[1].scatter(
    x=range(len(data)),
    y=data,
    color="lime",
    marker="p",  # marker shape
    s=200,  # marker size
)


axs[2].set_title("Step plot")
steps = axs[2].step(
    x=range(len(data)),
    y=data,
    color="aqua",
    where="mid",
    linewidth=3,
)
```

## Essential methods

### Example 1

```python
fig, ax = plt.subplots(
    figsize=(16, 8)
)

data = [4, 2, 5, 1, 1, 4]

ax.set_title("Scatter plot")
points = ax.scatter(
    x=range(len(data)),
    y=data,
    color="lime",
    marker="p",  # marker shape
    s=200,  # marker size
    label="my data",
)

ax.set_xlabel("the X")
ax.set_ylabel("the Y")

ax.legend()
ax.grid()
```

### Example 2 - axis limits and ticks spacing

```python
fig, ax = plt.subplots(
    figsize=(16, 8)
)

data = [4, 2, 5, 1, 1, 4]

ax.set_title("Scatter plot")
points = ax.scatter(
    x=range(len(data)),
    y=data,
    color="lime",
    marker="p",  # marker shape
    s=200,  # marker size
    label="my data",
)

ax.set_xlabel("the X")
ax.set_ylabel("the Y")

ax.legend()
ax.grid()

ax.set_xlim(-1, 7)
ax.set_ylim(0, 6)

ax.set_yticks([0, 2, 4, 6])
```

### Example 3 - grid behind plots

```python
fig, ax = plt.subplots(
    figsize=(16, 8)
)

data = [4, 2, 5, 1, 1, 4]

ax.grid()

ax.set_title("Scatter plot")
points = ax.scatter(
    x=range(len(data)),
    y=data,
    color="lime",
    marker="p",  # marker shape
    s=200,  # marker size
    label="my data",
)

ax.set_xlabel("the X")
ax.set_ylabel("the Y")

ax.legend()


ax.set_xlim(-1, 7)
ax.set_ylim(0, 6)

ax.set_yticks([0, 2, 4, 6])

ax.set_axisbelow(True)
```

## Figure titles and axes titles

```python
fig, axs = plt.subplots(
    1, 3,
    figsize=(16, 8)
)

fig.suptitle(
    "The same data plotted as...",
    size=20,
    weight="bold",
)

data = [4, 2, 5, 1, 1, 4]

axs[0].set_title("Bar plot")
bars = axs[0].bar(
    x=range(len(data)),
    height=data,
    tick_label=list("ABCDEF"),
    color="magenta"
)


axs[1].set_title("Scatter plot")
points = axs[1].scatter(
    x=range(len(data)),
    y=data,
    color="lime",
    marker="p",  # marker shape
    s=200,  # marker size
)


axs[2].set_title("Step plot")
steps = axs[2].step(
    x=range(len(data)),
    y=data,
    color="aqua",
    where="mid",
    linewidth=3,
)
```

## Annotations

```python
fig, ax = plt.subplots(
    figsize=(16, 8)
)

data = [4, 2, 5, 1, 1, 4]

ax.set_title("Points and steps")
points = ax.scatter(
    x=range(len(data)),
    y=data,
    color="lime",
    marker="p",  # marker shape
    s=200,  # marker size
)

ax.text(
    x=1.4, y=4.95,
    s="This is MAX",
    size=14,
    color="lime"
)

ax.annotate(
    text="This is MIN",
    xy=(3, 1),  # point I'm annotating
    xytext=(2, 2),  # location of the annotation
    color="red",
    size=14,
    arrowprops=dict(
        arrowstyle="->",
        connectionstyle="arc3,rad=-0.2",
        color="red"
    )
)
```
