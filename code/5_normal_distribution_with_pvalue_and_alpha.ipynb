# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# Standard normal distirbution

##  Plot 1: basic graph

```python
# DATA
mu = 0  # mean
stdev = 1  # standard deviation

# `x` span of +/- 4 standard deviations
x_start = mu - 4*stdev
x_end = mu + 4*stdev
step = 0.01

# Calculate `x` and `y`
x = np.arange(
    start=x_start,
    stop=x_end+step,
    step=step
)
y = norm.pdf(
    x=x,
    loc=mu,
    scale=stdev
)
```

```python
# PLOTTING
fig, ax = plt.subplots(
    figsize=(16, 9),
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density"
);
```

## Plot 2: basic graph with shading

```python
# DATA
# from previous example

# PLOTTING
fig, ax = plt.subplots(
    figsize=(16, 9),
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density"
)
# plot: normal distribution fill
ax.fill_between(
    x, y,
    alpha=0.2,
);
```

## Plot 3: basic graph with shading and tweaked axes appearance

```python
# DATA
# from previous example

# PLOTTING
fig, ax = plt.subplots(
    figsize=(16, 9),
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density"
)
# plot: normal distribution fill
ax.fill_between(
    x, y,
    alpha=0.2,
)

# SETTINGS
# axis
ax.set_xlim(x_start, x_end)
ax.set_xticks(np.arange(x_start, x_end+1, stdev)[1:-1])
ax.set_ylim(0, 0.45)
ax.set_yticks(np.arange(0, 0.46, 0.05)[1:-1])
ax.tick_params(
    axis="both",
    labelsize=16
)

# labels
ax.set_title(
    "Normal distribution plot",
    fontsize=20
)
ax.set_xlabel(
    "Z-score",
    fontsize=16
)
ax.set_ylabel(
    "Probability density",
    fontsize=16
);
```

# Real life data example

## Plot 4: normal distribution of height

```python
# DATA
mu = 183  # mean male height in Netherlands according to WHO
stdev = 7.1  # standard deviation

x_start = mu - 4*stdev
x_end = mu + 4*stdev
step = 0.01

# Calculate "x" and "y"
x = np.arange(
    start=x_start,
    stop=x_end+step,
    step=step
)
y = norm.pdf(
    x=x,
    loc=mu,
    scale=stdev
)
```

```python
# PLOTTING
fig, ax = plt.subplots(
    figsize=(16, 9)
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density"
)
# plot: normal distribution fill
ax.fill_between(
    x, y,
    alpha=0.2,
)

# SETTINGS
# axis
ax.set_xlim(x_start, x_end)
ax.set_xticks(np.arange(x_start, x_end+1, stdev)[1:-1])
ax.set_ylim(0, 0.060)
ax.set_yticks(np.arange(0, 0.065, 0.005)[1:-1])
ax.tick_params(
    axis="both",
    labelsize=16
)

# labels
ax.set_title(
    "Normal distribution plot of male population height in Netherlands",
    fontsize=20
)
ax.set_xlabel(
    "Height [cm]",
    fontsize=16
)
ax.set_ylabel(
    "Probability density",
    fontsize=16
);
```

## Plot 5: add mean value line

```python
# DATA
# from previous example

# PLOTTING
fig, ax = plt.subplots(
    figsize=(16, 9)
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density"
)
# plot: normal distribution fill
ax.fill_between(
    x, y,
    alpha=0.2,
)
# plot: mean line
ax.vlines(
    x=mu,
    ymin=0,
    ymax=max(y),
    linewidths=3.0,
    label="mean: {:.1f}".format(mu)
)

# SETTINGS
# axis
ax.set_xlim(x_start, x_end)
ax.set_xticks(np.arange(x_start, x_end+1, stdev)[1:-1])
ax.set_ylim(0, 0.060)
ax.set_yticks(np.arange(0, 0.065, 0.005)[1:-1])
ax.tick_params(
    axis="both",
    labelsize=16
)

# labels
ax.set_title(
    "Normal distribution plot of male population height in Netherlands",
    fontsize=20
)
ax.set_xlabel(
    "Height [cm]",
    fontsize=16
)
ax.set_ylabel(
    "Probability density",
    fontsize=16
)

# legend
ax.legend(
    fontsize=16,
    loc="upper left",
    frameon=False
);
```

## Plot 7: add two-sided p-value shading area

```python
# DATA
# from previous example

# PLOTTING
fig, ax = plt.subplots(
    figsize=(16, 9)
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density"
)
# plot: normal distribution fill
ax.fill_between(
    x, y,
    alpha=0.2,
)
# plot: mean line
ax.vlines(
    x=mu,
    ymin=0,
    ymax=max(y),
    linewidths=3.0,
    label="mean: {:.1f}".format(mu)
)
# plot: new observation
ax.scatter(
    x=x_observed,
    y=0,
    clip_on=False,
    zorder=10,
    label="E - new observation: {:.1f}".format(x_observed)
)
# plot: p-value
ax.fill_between(
    x, y,
    where=y < y_observed,
    label="p-value(E)",
)

# SETTINGS
# axis
ax.set_xlim(x_start, x_end)
ax.set_xticks(np.arange(x_start, x_end+1, stdev)[1:-1])
ax.set_ylim(0, 0.060)
ax.set_yticks(np.arange(0, 0.065, 0.005)[1:-1])
ax.tick_params(
    axis="both",
    labelsize=16
)

# labels
ax.set_title(
    "Normal distribution plot of male population height in Netherlands",
    fontsize=20
)
ax.set_xlabel(
    "Height [cm]",
    fontsize=16
)
ax.set_ylabel(
    "Probability density",
    fontsize=16
)

# legend
ax.legend(
    fontsize=16,
    loc="upper left",
    frameon=False
);
```

## Plot 8: cdf to explain alpha

**Note:** The significance level or alpha level is the probability of making the wrong decision when the null hypothesis is true. 

```python
# DATA
y_cdf = norm.cdf(
    x=x,
    loc=mu,
    scale=stdev
)

# significance level alpha (two-sided)
alpha = 0.05
left_alpha = alpha / 2
right_alpha = 1 - (alpha / 2)
```

```python
# DATA
# from previous example

# PLOTTING
fig, ax = plt.subplots(
    figsize=(16, 9)
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density"
)
# plot: normal distribution f
ax.fill_between(
    x, y,
    alpha=0.2,
)
# plot: mean line
ax.vlines(
    x=mu,
    ymin=0,
    ymax=max(y),
    linewidths=3.0,
    label="mean: {:.1f}".format(mu)
)
# plot: new observation
ax.scatter(
    x=x_observed,
    y=0,
    clip_on=False,
    zorder=10,
    label="E - new observation: {:.1f}".format(x_observed)
)
# plot: p-value
ax.fill_between(
    x, y,
    where=y < y_observed,
    label="p-value(E)",
)
# plot: cdf
ax2 = ax.twinx()
ax2.plot(
    x, y_cdf,
    color="orange",
    linewidth=1.75,
    label="cdf"
)

# SETTINGS
# axis
ax.set_xlim(x_start, x_end)
ax.set_xticks(np.arange(x_start, x_end+1, stdev)[1:-1])
ax.set_ylim(0, 0.060)
ax.set_yticks(np.arange(0, 0.065, 0.005)[1:-1])
ax.tick_params(
    axis="both",
    labelsize=16
)

ax2.set_yticks(np.arange(0, 1.05, 0.05)[1:])
ax2.set_yticklabels(
    [f"{y: .0%}" for y in np.arange(0, 1.05, 0.05)[1:]],
    color="orange"
)
ax2.set_ylim(0, 1.05)

# labels
ax.set_title(
    "Normal distribution plot of male population height in Netherlands",
    fontsize=20
)
ax.set_xlabel(
    "Height [cm]",
    fontsize=16
)
ax.set_ylabel(
    "Probability density",
    fontsize=16
)

ax2.set_ylabel(
    "CDF",
    fontsize=16,
    color="orange"
)

# legends
ax.legend(
    fontsize=16,
    loc="upper left",
    frameon=False
)
```

## Plot 9: add two-sided alpha

```python
# DATA
# from previous example

# PLOTTING
fig, ax = plt.subplots(
    figsize=(16, 9)
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density"
)
# plot: normal distribution fill
ax.fill_between(
    x, y,
    alpha=0.2,
)
# plot: mean line
ax.vlines(
    x=mu,
    ymin=0,
    ymax = max(y),
    linewidths=3.0,
    label="mean: {:.1f}".format(mu)
)
# plot: new observation
ax.scatter(
    x=x_observed,
    y=0,
    clip_on=False,
    zorder=10,
    label="E - new observation: {:.1f}".format(x_observed)
)
# plot: p-value
ax.fill_between(
    x, y,
    where=y < y_observed,
    label="p-value(E)",
)
# plot: alpha
ax.fill_between(
    x, y,
    where=(y_cdf <= left_alpha) | (y_cdf >= right_alpha),
    label="alpha: {}".format(alpha),
)

# SETTINGS
# axis
ax.set_xlim(x_start, x_end)
ax.set_xticks(np.arange(x_start, x_end+1, stdev)[1:-1])
ax.set_ylim(0, 0.060)
ax.set_yticks(np.arange(0, 0.065, 0.005)[1:-1])
ax.tick_params(
    axis="both",
    labelsize=16
)

# labels
ax.set_title(
    "Normal distribution plot of male population height in Netherlands",
    fontsize=20
)
ax.set_xlabel(
    "Height [cm]",
    fontsize=16
)
ax.set_ylabel(
    "Probability density",
    fontsize=16
)

# legend
ax.legend(
    fontsize=16,
    loc="upper left",
    frameon=False
);
```

## Plot 10: final result

```python
# PLOTTING
gnome_colour_1 = "#8ff0a4"  # GNOME green 1
gnome_colour_5 = "#26a269"  # GNOME green 5

fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white"
)

# plot: normal distribution
ax.plot(
    x, y,
    label="probability density",
    color=gnome_colour_5,
    linewidth=2,
)
# plot: normal distribution fill
ax.fill_between(
    x, y,
    color="white",
    alpha=0.2,
)
# plot: mean line
ax.vlines(
    x=mu,
    ymin=0,
    ymax=max(y),
    label="mean: {:.1f}".format(mu),
    color=gnome_colour_5,
    linestyle="--",
    linewidth=2,
)
# plot: new observation
ax.scatter(
    x=x_observed,
    y=0,
    clip_on=False,
    zorder=10,
    label="E - new observation: {:.1f}".format(x_observed),
    marker="v",
    s=125,  # size
    color="white",
    edgecolor=gnome_colour_5,
    linewidth=2
)
# plot: p-value
ax.fill_between(
    x, y,
    where=y < y_observed,
    label="p-value(E)",
    facecolor=gnome_colour_1,
    edgecolor="none",
)
# plot: alpha
ax.fill_between(
    x, y,
    where=(y_cdf <= left_alpha) | (y_cdf >= right_alpha),
    label = "alpha: {}".format(alpha),
    facecolor="none",
    edgecolor=gnome_colour_5,
    hatch="o",
    linewidth=0
)

# SETTINGS
# axis
ax.set_xlim(x_start, x_end)
ax.set_xticks(np.arange(x_start, x_end+1, stdev)[1:-1])
ax.set_ylim(0, 0.060)
ax.set_yticks(np.arange(0, 0.065, 0.005)[1:-1])
ax.tick_params(
    axis="both",
    labelsize=16
)

# labels
ax.set_title(
    "Normal distribution plot of male population height in Netherlands",
    fontsize=20
)
ax.set_xlabel(
    "Height [cm]",
    fontsize=16
)
ax.set_ylabel(
    "Probability density",
    fontsize=16
)

# legend
ax.legend(
    fontsize=16,
    loc="upper left",
    frameon=False
);

plt.tight_layout()
plt.savefig(
    fname="normal_plot.png",
    dpi=125,
    facecolor=[1, 1, 1],
);
```
