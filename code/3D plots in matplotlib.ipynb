# Create empty 3D figure

In order to create any plot using `matplotlib` we always follow the same steps. First, we create a `figure` object and then we create `axes` object.

```python
# 2D
fig = plt.figure(facecolor='palegreen')
ax = fig.add_subplot()
ax.set_title(type(ax));
```

There are 3 ways to create an axes:

* [fig.add_subplot](https://matplotlib.org/stable/api/figure_api.html?highlight=add_subplot#matplotlib.figure.Figure.add_subplot)
* [fig.add_axes()](https://matplotlib.org/stable/api/figure_api.html?highlight=figure#matplotlib.figure.Figure.add_axes)
* [fig.subplots()](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.subplots)

### Create empty 3D `axis` using `projection` method

When creating a 3D plot, we follow the same workflow of creating the `figure` and then `axis`. However, for the 3D plot, we need to specify the `projection` parameter when creating an `axis`.

* [fig.add_subplot](https://matplotlib.org/stable/api/figure_api.html?highlight=add_subplot#matplotlib.figure.Figure.add_subplot)

```python
# 3d
fig = plt.figure(facecolor='palegreen')
ax = fig.add_subplot(projection='3d')
ax.set_title(type(ax));
```

**Note:** the value of `projection` must be lowercase '3d'. Upper case '3D' will produce an error.

* [fig.add_axes()](https://matplotlib.org/stable/api/figure_api.html?highlight=figure#matplotlib.figure.Figure.add_axes)

```python
fig = plt.figure(facecolor='palegreen')
ax = fig.add_axes([0, 0 , 1, 1],
                  projection='3d'
                 )
ax.set_title(type(ax));
```

* [plt.subplots()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib-pyplot-subplots)

```python
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.set_title(type(ax));
```

# Types of plot

All of the examples below we'll be using the `fig.add_subplot()` method for creating an axis

## Scatter
### Scatter with 1 data point

```python
fig = plt.figure(figsize=(16, 9),
                 facecolor='palegreen'
                )
ax = fig.add_subplot(projection='3d')
ax.scatter(0.5, 0.5, 0.5)

# Settings
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_title('Scatter with 1 data point')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z_axis');
```

### Scatter with multiple data points: example 1

```python
# Data
x = [0.25, 0.5, 0.75]
y = [0.25, 0.5, 0.75]
z = [0.25, 0.5, 0.75]

fig = plt.figure(figsize=(16, 9),
                 facecolor='palegreen'
                )
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z)

# Settings
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_title('Scatter with multiple data points')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z_axis');
```

It can be noticed that some points are faded away. To remove this effect, we need to set the `depth shade` paramter to `False`:

```python
fig = plt.figure(figsize=(16, 9),
                 facecolor='palegreen'
                )
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z,
           depthshade=False)

# Settings
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_title('Scatter with multiple data points')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z_axis');
```

### Scatter with multiple data points: example 1
3D scatter plot will take any argument a regular 2D scatter plot would take. For more options, see:

* [ax.scatter()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html#matplotlib-axes-axes-scatter)

Let's create a plot with more data points and customize them a bit:

```python
# Data
np
x = np.random.random(30)
y = np.random.random(30)
z = np.random.random(30)

fig = plt.figure(figsize=(16, 9),
                 facecolor='palegreen'
                )
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z,
           c='g',
           edgecolor='r',
           linewidth=2,
           s=100,
           marker='D'
           )

# Settings
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_title('Scatter with multiple data points')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z_axis');
```
