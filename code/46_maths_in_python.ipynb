## 4) `decimal`

```python
from decimal import Decimal, getcontext
```

Binary floating point problem:

```python
a = 0.15
a
```

```python
b = 0.25
```

```python
# It seems OK
a + b
```

```python
# But in reality:
print(f"{a + b: .20f}")
```

```python
c = 0.016
c
```

```python
# You don't even have to increase precision to sometimes observe the floating point error
a + b + c
```

Decimal representation:

```python
A = Decimal("0.15")
A
```

```python
B = Decimal("0.25")
B
```

```python
# It seems OK
A + B
```

```python
# And it is OK
print(f"{A + B: .20f}")
```

```python
C = Decimal("0.016")
C
```

```python
A + B + C
```

```python
print(f"{A + B + C: .20f}")
```

See default settings.

```python
getcontext()
```

Change default precision

```python
Decimal("1") / Decimal("6")
```

```python
# Update precision
getcontext().prec = 4
```

```python
Decimal("1") / Decimal("6")
```

Comparison

```python
Decimal("0.1234") == Decimal("0.1234")
```

```python
Decimal("0.1234") < 0.13
```

Math functions

```python
Decimal(9).sqrt()
```

Rounding

```python
Decimal("2.1").quantize(Decimal("1"), rounding="ROUND_UP")
```

## 7) `statistics`

```python
import statistics
```

```python
data = [1,3,10,13,18,20,28,29]
data
```

```python
statistics.mean(data)
```

```python
# For the sample
statistics.stdev(data)
```

```python
# For the population
statistics.pstdev(data)
```

```python
statistics.quantiles(data)
```

```python
statistics.linear_regression(range(len(data)), data)
```

```python
statistics.NormalDist(175, 5)  # (mean, stdev)
```

```python
statistics.NormalDist(175, 5).samples(5)
```

```python
statistics.NormalDist(175, 5).zscore(170)
```
