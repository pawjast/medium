# `matplotlib` R style bar chart

## Imports

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
from cycler import cycler
```

## Data

```python
x = list("ABCDE")
y = [3, 5, 1, 4, 2]
```

## Plot v1: native R's barplot look

### Default matplotlib look

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("Default matplotlib bar chart")

ax.bar(
    x=x,
    height=y,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")
```

### Step 1: remove spines

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("R style bar chart in the making...")

ax.bar(
    x=x,
    height=y,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

# REMOVE SPINES
# -------------
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)
```

### Step 2: remove ticks

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("R style bar chart in the making...")

ax.bar(
    x=x,
    height=y,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

# REMOVE SPINES
# -------------
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)


# REMOVE X TICKS
# --------------
ax.tick_params(
    axis="x", 
    bottom=False, # This makes X axis ticks hidden
)
```

### Step 3: update the y axis

* Make the ticks longer
* make sure the spine starts and end with a tick

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("R style bar chart in the making...")

ax.bar(
    x=x,
    height=y,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

# REMOVE SPINES
# -------------
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)


# REMOVE X TICKS
# --------------
ax.tick_params(
    axis="x", 
    bottom=False, # This makes X axis ticks hidden
)

# UPDATE Y AXIS
# -------------
ax.set_ylim(0, 5)

ax.tick_params(
    axis="y",
    length=10,  # Make ticks longer
)
```

### Step 4: change bar colours

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("R style bar chart in the making...")

ax.bar(
    x=x,
    height=y,
    facecolor="silver",
    edgecolor="black",
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

# REMOVE SPINES
# -------------
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)


# REMOVE X TICKS
# --------------
ax.tick_params(
    axis="x", 
    bottom=False, # This makes X axis ticks hidden
)

# UPDATE Y AXIS
# -------------
ax.set_ylim(0, 5)

ax.tick_params(
    axis="y",
    length=10,  # Make ticks longer
)
```

### Step 5: disable clipping

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("R style bar chart")

ax.bar(
    x=x,
    height=y,
    facecolor="silver",
    edgecolor="black",
    clip_on=False,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

# REMOVE SPINES
# -------------
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)


# REMOVE X TICKS
# --------------
ax.tick_params(
    axis="x", 
    bottom=False, # This makes X axis ticks hidden
)

# UPDATE Y AXIS
# -------------
ax.set_ylim(0, 5)

ax.tick_params(
    axis="y",
    length=10,  # Make ticks longer
)
```

## Plot v2

### Step 1: add bg colour

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("R style bar chart")

ax.bar(
    x=x,
    height=y,
    facecolor="silver",
    edgecolor="black",
    clip_on=False,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

# REMOVE SPINES
# -------------
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)


# REMOVE X TICKS
# --------------
ax.tick_params(
    axis="x", 
    bottom=False, # This makes X axis ticks hidden
)

# UPDATE Y AXIS
# -------------
ax.set_ylim(0, 5)

ax.tick_params(
    axis="y",
    length=10,  # Make ticks longer
)

# ADD BACKGROUND COLOUR
# ---------------------
ax.set_facecolor("whitesmoke")
```

### Step 2: offset the Y axis

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("R style bar chart")

ax.bar(
    x=x,
    height=y,
    facecolor="silver",
    edgecolor="black",
    clip_on=False,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

# REMOVE SPINES
# -------------
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)


# REMOVE X TICKS
# --------------
ax.tick_params(
    axis="x", 
    bottom=False, # This makes X axis ticks hidden
)

# UPDATE Y AXIS
# -------------
ax.set_ylim(0, 5)

ax.tick_params(
    axis="y",
    length=10,  # Make ticks longer
)

# ADD BACKGROUND COLOUR
# ---------------------
ax.set_facecolor("whitesmoke")

# OFFSET THE Y AXIS
# ------------------
ax.spines["left"].set_position(("outward", 15))
```

### Step 3: use rcParams

```python
common_width = 1.75

rcParams.update(
    {
        "font.size": 18,
        "axes.titlesize": 25,
        "grid.linewidth": common_width,
        "axes.linewidth": common_width,  # Y spine thickness
        "ytick.major.width": common_width,
        "patch.linewidth": 1.5,  # thicknes of the bar edge
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(16, 9),
    facecolor="white",
    layout="tight",

)

fig.suptitle("R style bar chart")

ax.bar(
    x=x,
    height=y,
    facecolor="silver",
    edgecolor="black",
    clip_on=False,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

# REMOVE SPINES
# -------------
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)


# REMOVE X TICKS
# --------------
ax.tick_params(
    axis="x", 
    bottom=False, # This makes X axis ticks hidden
)

# UPDATE Y AXIS
# -------------
ax.set_ylim(0, 5)

ax.tick_params(
    axis="y",
    length=10,  # Make ticks longer
)

# ADD BACKGROUND COLOUR
# ---------------------
ax.set_facecolor("whitesmoke")

# OFFSET THE Y AXIS
# ------------------
ax.spines["left"].set_position(("outward", 15))
```

### Step 3: expand rcParams

```python
common_width = 1.75

rcParams.update(
    {
        "font.size": 18,
        "axes.titlesize": 25,
        "grid.linewidth": common_width,
        "axes.linewidth": common_width,  # Y spine thickness
        "ytick.major.width": common_width,
        "patch.linewidth": 1.5,  # thicknes of the bar edge

        # Figure Settings
        "figure.figsize": (16, 9),
        "figure.facecolor": "white",
        "figure.autolayout": True,  # Equivelant to layout="tight"

        # Show/hide spines
        "axes.spines.top": False,
        "axes.spines.right": False,
        "axes.spines.bottom": False,
        "axes.spines.left": True,

        # Ticks
        "xtick.bottom": False,  # Hide X ticks
        "ytick.major.size": 10,  # Make Y ticks longer

        # Update Axes bgcolor
        "axes.facecolor": "whitesmoke",

        # Update bar facecolour and edge
        "patch.edgecolor": "black",
        "patch.force_edgecolor": True,  # Forces bar edges to respect patch.edgecolor

        "axes.prop_cycle": cycler(color=["silver"]),
    }
)
```

```python
fig, ax = plt.subplots()

fig.suptitle("R style bar chart")

p = ax.bar(
    x=x,
    height=y,
    clip_on=False,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

ax.set_ylim(0, 5)

# OFFSET THE Y AXIS
# ------------------
ax.spines["left"].set_position(("outward", 15))
```

## Plot v3: ggplot look

```python
common_width = 1.75

rcParams.update(
    {
        "font.size": 18,
        "axes.titlesize": 25,
        "grid.linewidth": common_width,
        "axes.linewidth": common_width,  # Y spine thickness
        "ytick.major.width": common_width,
        "xtick.major.width": common_width,
        "patch.linewidth": 1.5,  # thicknes of the bar edge

        # Figure Settings
        "figure.figsize": (16, 9),
        "figure.facecolor": "white",
        "figure.autolayout": True,  # Equivelant to layout="tight"

        # Show/hide spines
        "axes.spines.top": False,
        "axes.spines.right": False,
        "axes.spines.bottom": False,
        "axes.spines.left": False,

        # Ticks
        "xtick.bottom": True,  # Show X ticks
        "ytick.major.size": 10,  # Make Y ticks longer
        "xtick.major.size": 10,  # Make X ticks longer

        # Update Axes bgcolor
        "axes.facecolor": "whitesmoke",

        # Update bar facecolour and edge
        "patch.edgecolor": "none",
        "patch.force_edgecolor": True,  # Forces bar edges to respect patch.edgecolor

        "axes.prop_cycle": cycler(color=["dimgrey"]),

        # Add gridline
        "axes.grid": True,
        "grid.color": "white",
        "grid.linewidth": common_width,
        "axes.axisbelow": True, # Keep grid behind plots
    }
)
```

```python
fig, ax = plt.subplots()

fig.suptitle("R style ggplot2 bar chart")

p = ax.bar(
    x=x,
    height=y,
    clip_on=False,
)

ax.set_xlabel("Categories")
ax.set_ylabel("Count")

ax.set_ylim(-0.25, 5.25)
```
