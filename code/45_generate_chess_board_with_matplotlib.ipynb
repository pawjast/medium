# `matplotlib` plot of a chessboard

20th July - International chess day

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Base figure

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="lightgrey",
    layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
);
```

## Data

```python
odd_row = np.array([bool(x) for x in [1, 0, 1, 0, 1, 0, 1, 0]])
print(odd_row)
```

```python
# even_row
~odd_row
```

## Stacked bar charts - first row

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="lightgrey",
    layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Plot stacked bars
ax.bar(
    x=list("ABCDEFGH"),
    height=odd_row,
    color="black",
    width=1,
)
ax.bar(
    x=list("ABCDEFGH"),
    height=~odd_row,
    color="black",
    width=1,
    bottom=np.ones(len(odd_row))
);
```

## Stacked bar - more space

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="lightgrey",
    layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Plot stacked bars
ax.bar(
        x=list("ABCDEFGH"),
        height=odd_row,
        color="black",
        width=1,
    )
ax.bar(
    x=list("ABCDEFGH"),
    height=~odd_row,
    color="black",
    width=1,
    bottom=np.ones(len(odd_row))
)

ax.set_xlim(-0.5, 7.5)
ax.set_ylim(0, 8)
ax.set_yticks(
    ticks=np.arange(0.5, 8, 1),
    labels=[f"{y}" for y in range(1, 9)]
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
# );
```

## Stacked bar charts - FINAL

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="lightgrey",
    layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Plot stacked bars
for i in range(0, 8, 2):
    ax.bar(
        x=list("ABCDEFGH"),
        height=odd_row,
        color="black",
        width=1,
        bottom=np.ones(len(odd_row))*i
    )
    ax.bar(
        x=list("ABCDEFGH"),
        height=~odd_row,
        color="black",
        width=1,
        bottom=np.ones(len(odd_row))*(i+1)
    )

ax.set_xlim(-0.5, 7.5)
ax.set_ylim(0, 8)
ax.set_yticks(
    ticks=np.arange(0.5, 8, 1),
    labels=[f"{y}" for y in range(1, 9)]
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
# );
```

## Scatter plot - data

```python
# Generate mesh of data
xv_odd, yv_odd  = np.meshgrid(range(1, 8, 2), range(1, 8, 2))
```

```python
xv_odd
```

```python
yv_odd
```

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="lightgrey",
    layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Plot scatter plot


ax.scatter(
    xv_odd, yv_odd
)
ax.scatter(
    xv_odd+1, yv_odd+1
)

ax.set_xlim(0.5, 8.5)
ax.set_ylim(0.5, 8.5)
ax.set_xticks(
    ticks=np.arange(1, 9, 1),
    labels=list("ABCDEFGH")
)
ax.set_yticks(
    ticks=np.arange(1, 9, 1),
    labels=[f"{y}" for y in range(1, 9)]
);
```

## Scatter plot - FINAL

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="lightgrey",
    layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Plot scatter plot

kwargs = {
    "marker": "s",
    "s": 3750,  # marker size
    "c": "black"  # marker colour
}

ax.scatter(
    xv_odd, yv_odd, **kwargs
)
ax.scatter(
    xv_odd+1, yv_odd+1, **kwargs
)

ax.set_xlim(0.5, 8.5)
ax.set_ylim(0.5, 8.5)
ax.set_xticks(
    ticks=np.arange(1, 9, 1),
    labels=list("ABCDEFGH")
)
ax.set_yticks(
    ticks=np.arange(1, 9, 1),
    labels=[f"{y}" for y in range(1, 9)]
);


# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
# );
```

## Subplots - new base figure

```python
fig, axs = plt.subplots(
    8, 8,
    figsize=(8, 8),
    facecolor="lightgrey",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.subplots_adjust(
    hspace=0,
    wspace=0
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
);
```

## Subplots - add colours

```python
fig, axs = plt.subplots(
    8, 8,
    figsize=(8, 8),
    facecolor="lightgrey",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.subplots_adjust(
    hspace=0,
    wspace=0
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Colour black alternate subplots
for x in range(0, 8, 2):
    for y in range(1, 9, 2):
            axs[x, y].set_facecolor("black")
for x in range(1, 9, 2):
    for y in range(0, 8, 2):
            axs[x, y].set_facecolor("black")

# Remove all ticks from all plots
for a in axs.flat:
    a.set_xticks([])
    a.set_yticks([])
```

## Subplots - final

```python
fig, axs = plt.subplots(
    8, 8,
    figsize=(8, 8),
    facecolor="lightgrey",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.subplots_adjust(
    hspace=0,
    wspace=0
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Colour black alternate subplots
for x in range(0, 8, 2):
    for y in range(1, 9, 2):
            axs[x, y].set_facecolor("black")
for x in range(1, 9, 2):
    for y in range(0, 8, 2):
            axs[x, y].set_facecolor("black")

# Remove all ticks from all plots
for a in axs.flat:
    a.set_xticks([])
    a.set_yticks([])

# Add ticks and labels to the bottom row
for a, t in zip(axs[-1, :], list("ABCDEFGH")):
    a.set_xticks(
            ticks=[0.5],
            labels=t
    )

# Add ticks and labels to the first column
for a, t in zip(axs[:, 0], range(8,0, -1)):
    a.set_yticks(
            ticks=[0.5],
            labels=f"{t}"
    )


# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
# );
```
