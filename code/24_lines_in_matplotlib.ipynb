## Axes lines

* It can only plot one line at a time
* Line value expressed in data terms
* Line limits expressed in axis terms [0, 1]

### `ax.axhline()`

Horizontal lines.

```python
colour_val = "#e01b24"  # GNOME Red 3

fig, axs = plt.subplots(
    1, 3,
    sharex=True,
    sharey=True,
    layout="constrained"
)
fig.suptitle("ax.axhline()")

# Set limits
axs[0].set_xlim(0, 10)
axs[0].set_ylim(0, 10)

# Plots
axs[0].set_title("A")
axs[0].axhline(
    y=5,  # data coordinates
    color=colour_val
)

axs[1].set_title("B")
axs[1].axhline(
    y=3,  # data coordinates
    xmax=.5,  # axe coordinates [0, 1]
    color=colour_val
)
axs[1].axhline(
    y=7,  # data coordinates
    xmin=.5,  # axe coordinates [0, 1]
    color=colour_val
)

axs[2].set_title("C")
axs[2].axhline(
    y=1,
    xmin=.2,  # axe coordinates [0, 1]
    xmax=.8,  # axe coordinates [0, 1]
    color=colour_val
)
axs[2].axhline(
    y=5,
    xmin=.3,  # axe coordinates [0, 1]
    xmax=.7,  # axe coordinates [0, 1]
    color=colour_val
)
axs[2].axhline(
    y=9,
    xmin=.4,  # axe coordinates [0, 1]
    xmax=.6,  # axe coordinates [0, 1]
    color=colour_val
)


for ax in axs:
    ax.set_xticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{x}" for x in np.arange(2, 10, 2)] + [""],
    )
    ax.set_yticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{y}" for y in np.arange(2, 10, 2)] + [""],
    )
    ax.grid()
```

### `ax.axvline()`

Vertical lines.

```python
colour_val = "#9141ac"  # GNOME Purple 3

fig, axs = plt.subplots(
    1, 3,
    sharex=True,
    sharey=True,
    layout="constrained"
)
fig.suptitle("ax.axvline()")

# Set limits
axs[0].set_xlim(0, 10)
axs[0].set_ylim(0, 10)

# Plots
axs[0].set_title("D")
axs[0].axvline(
    x=5,
    color=colour_val
)

axs[1].set_title("E")
axs[1].axvline(
    x=3,  # data coordinates
    ymax=.5,  # axe coordinates [0, 1]
    color=colour_val
)
axs[1].axvline(
    x=7,  # data coordinates
    ymin=.5,  # axe coordinates [0, 1]
    color=colour_val
)

axs[2].set_title("F")
axs[2].axvline(
    x=1,  # data coordinates
    ymin=.2,  # axe coordinates [0, 1]
    ymax=.8,  # axe coordinates [0, 1]
    color=colour_val
)
axs[2].axvline(
    x=5,  # data coordinates
    ymin=.3,  # axe coordinates [0, 1]
    ymax=.7,  # axe coordinates [0, 1]
    color=colour_val
)
axs[2].axvline(
    x=9,  # data coordinates
    ymin=.4,  # axe coordinates [0, 1]
    ymax=.6,  # axe coordinates [0, 1]
    color=colour_val
)


for ax in axs:
    ax.set_xticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{x}" for x in np.arange(2, 10, 2)] + [""],
    )
    ax.set_yticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{y}" for y in np.arange(2, 10, 2)] + [""],
    )
    ax.grid()
```

### `axline()`

Angular lines.

```python
colour_val = "#ff7800"  # GNOME Orange 3

fig, axs = plt.subplots(
    1, 2,
    sharex=True,
    sharey=True,
    layout="constrained"
)
fig.suptitle("ax.axline()")

# Set limits
axs[0].set_xlim(0, 10)
axs[0].set_ylim(0, 10)

# Plots
axs[0].set_title("G")
axs[0].axline(
    xy1=(0, 0),
    xy2=(1, 1),
    color=colour_val
)

axs[1].set_title("H")
axs[1].axline(
    xy1=(5, 5), 
    slope=-1,
    color=colour_val
)

for ax in axs:
    ax.set_xticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{x}" for x in np.arange(2, 10, 2)] + [""],
    )
    ax.set_yticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{y}" for y in np.arange(2, 10, 2)] + [""],
    )
    ax.grid()
```

## Axes spans

### `ax.axhspan()`

```python
colour_val = "#3584e4"  # GNOME Blue 3

fig, axs = plt.subplots(
    1, 2,
    sharex=True,
    sharey=True,
    layout="constrained"
)
fig.suptitle("ax.axhspan()")

# Set limits
axs[0].set_xlim(0, 10)
axs[0].set_ylim(0, 10)

# Plots
axs[0].set_title("I")
axs[0].axhspan(
    ymin=3,  # data coordinates
    ymax=7,  # data coordinates
    color=colour_val
)

axs[1].set_title("J")
axs[1].axhspan(
    ymin=1,  # data coordinates
    ymax=2,  # data coordinates
    xmax=0.2,  # axe coordinates [0, 1]
    color=colour_val
)
axs[1].axhspan(
    ymin=4,  # data coordinates
    ymax=5.1,  # data coordinates
    xmin=0.4,  # axe coordinates [0, 1]
    xmax=0.65,  # axe coordinates [0, 1]
    color=colour_val
)
axs[1].axhspan(
    ymin=7,  # data coordinates
    ymax=9,  # data coordinates
    xmin=0.8,  # axe coordinates [0, 1]
    color=colour_val
)

for ax in axs:
    ax.set_xticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{x}" for x in np.arange(2, 10, 2)] + [""],
    )
    ax.set_yticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{y}" for y in np.arange(2, 10, 2)] + [""],
    )
    ax.grid()
```

## Lines in data coordinates

### `ax.hlines()`

```python
colour_val = "#e5a50a"  # GNOME Yellow 5

fig, axs = plt.subplots(
    1, 3,
    layout="constrained"
)
fig.suptitle("ax.hlines()")

# Set limits
axs[0].set_xlim(0, 10)
axs[0].set_ylim(0, 10)

# Plots
axs[0].set_title("M")
axs[0].hlines(
    y=[2, 4, 6, 8],  # data coordinates
    xmin=1,  # data coordinates
    xmax=9,  # data coordinates
    color=colour_val
)

axs[1].set_title("N")
axs[1].hlines(
    y=[2, 4, 6, 8],  # data coordinates
    xmin=1,  # data coordinates
    xmax=[9, 7, 5, 3],  # data coordinates
    color=colour_val
)

axs[2].set_title("O")
axs[2].hlines(
    y=[2, 4, 6, 8],  # data coordinates
    xmin=[1, 5, 1, 5],  # data coordinates
    xmax=[3, 7, 2, 9],  # data coordinates
    color=["red", "green", "blue", "black"]
)


for ax in axs:
    ax.set_xticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{x}" for x in np.arange(2, 10, 2)] + [""],
    )
    ax.set_yticks(
        ticks=np.arange(2, 11, 2),
        labels=[f"{y}" for y in np.arange(2, 10, 2)] + [""],
    )
    ax.grid()
```

## Real life example

### Generate data

```python
nr_of_points = 50

np.random.seed(2)
y_values = np.random.normal(
    loc=9,  # mean
    scale=2.5,  # std
    size=nr_of_points
)

y_values
```

```python
x_values = np.arange(nr_of_points)
x_values
```

### Calculations

```python
y_mean = np.mean(y_values)
y_std = np.std(y_values)
# Min and max values indexes
i_max = np.argmax(y_values)
i_min = np.argmin(y_values)
```

### Plot

```python
fig, ax = plt.subplots(
    1, 1,
    figsize=(18, 10),
    layout="constrained"
)
fig.suptitle("Randomly generated normally distributed set of numbers")

ax.set_xlim(-0.5, nr_of_points)
ax.set_ylim(0, 20)
ax.set_xlabel("X")
ax.set_ylabel("Y")

ax.plot(
    y_values,
    marker="o",
    markeredgecolor="blue",
    markerfacecolor="white",
    markeredgewidth=2,
    lw=0,
    label="data"
)

ax.axhline(
    y_mean,
    color="blue",
    lw=2,
    label="mean"
)

ax.axhspan(
    ymin=y_mean-y_std,
    ymax=y_mean+y_std,
    color="blue",
    alpha=0.1,
    label="+/- 1 std"
)

ax.hlines(
    y=max(y_values),
    xmin=-0.5,
    xmax=x_values[i_max],
    lw=0.75,
    label="max value",
    color="green"
)

ax.hlines(
    y=min(y_values),
    xmin=-0.5,
    xmax=x_values[i_min],
    lw=0.75,
    label="min value",
    color="red"
)

ax.legend(
    frameon=False,
    ncols=2,
    loc="upper right"
)

ax.set_yticks(
    ticks=[0, np.min(y_values), y_mean-y_std, y_mean, y_mean+y_std, np.max(y_values), 20]
);
```
