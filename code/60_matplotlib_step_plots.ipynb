# `matplotlib` step plots

## Imports

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

## Figure settings

```python
rcParams.update(
    {
        "figure.figsize": (16, 10),
        "figure.facecolor": "#FFFFFF",
        "figure.titlesize": 30,
        "figure.titleweight": "bold",
        "figure.constrained_layout.use": True,
        "font.size": 20,
        "lines.linewidth": 2.5,
        "grid.linewidth": 2,
        "grid.color": "#ffffff",
        "xtick.bottom": False,
        "ytick.left": False,
        "axes.facecolor": "#f5f5f5",
        "axes.spines.right": False,
        "axes.spines.left": False,
        "axes.spines.top": False,
        "axes.spines.bottom": False,
        "axes.axisbelow": True,  # grid behind the data
    }
)
```

## Data

```python
x = [1, 4, 10, 12, 15, 20]
y = [2, 4, 1, 3, 1, 4]
```

```python
fig, ax = plt.subplots()
fig.suptitle("Raw data")

# Plot the points
ax.scatter(
    x,
    y,
    marker="X",
    s=275,
    linewidth=1.5,
    edgecolor="black",
    facecolor="white",
    label="Data"
)

# Set axis limits
ax.set_ylim(0, 5.5)
ax.set_yticks(range(1, 6))
ax.set_xlim(0, 21)
ax.set_xticks(range(1, 21))

# Add legend
ax.legend(
    loc="upper center",
    facecolor="white",
    edgecolor="none"
)

# Add grid
ax.grid()

```

## ax.step()

### ax.step() - default

```python
fig, ax = plt.subplots()
fig.suptitle("ax.step(x,y) - default setting")

# Plot the points
ax.scatter(
    x,
    y,
    marker="X",
    s=275,
    linewidth=1.5,
    edgecolor="black",
    facecolor="white",
    zorder=100,  # crude way to keep the points always on top
    label="Data",
)

# Default step: `pre`
ax.step(
    x, y,
    color="red",
    label="ax.step()"
)

# Set axis limits
ax.set_ylim(0, 5.5)
ax.set_yticks(range(1, 6))
ax.set_xlim(0, 21)
ax.set_xticks(range(1, 21))

# Add legend
ax.legend(
    loc="upper center",
    facecolor="white",
    edgecolor="none"
)

# Add grid
ax.grid()
```

### ax.step() - all step options

```python
fig, axs = plt.subplots(
    3, 1,
    sharex=True,
    sharey=True
)
fig.suptitle("ax.step(x,y)")

# Plot the points
for ax in axs:
    ax.scatter(
        x, y,
        marker="X",
        s=275,
        linewidth=1.5,
        edgecolor="black",
        facecolor="white",
        zorder=100  # crude way to keep the points always on top
    )


# Plot the step lines
# Default step: `pre`
axs[0].step(
    x, y,
    color="red",
    label="ax.step(...)"
)
axs[1].step(
    x, y,
    color="blue",
    where="mid",
    label="ax.step(..., where='mid')"
)
axs[2].step(
    x, y,
    color="green",
    where="post",
    label="ax.step(..., where='post')"
)

# Set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(0, 21)
axs[2].set_xticks(range(1, 21))

# Add legend and grid
for ax in axs:
    ax.legend(
        loc="upper center",
        facecolor="white",
        edgecolor="none"
    )
    ax.grid()
```

## ax.plot()

### ax.plot() - a single option

```python
fig, ax = plt.subplots()
fig.suptitle("step plot with ax.plot(x, y)")

# Plot the points
ax.scatter(
    x, y,
    marker="X",
    s=275,
    linewidth=1.5,
    edgecolor="black",
    facecolor="white",
    zorder=100,  # crude way to keep the points always on top
    label="Data",
)

# Step plot with plot
ax.plot(
    x, y, 
    color="red",
    drawstyle="steps-pre",
    label="ax.plot(..., drawstyle='steps-pre')"
)

# Set axis limits
ax.set_ylim(0, 5.5)
ax.set_yticks(range(1, 6))
ax.set_xlim(0, 21)
ax.set_xticks(range(1, 21))

# Add legend
ax.legend(
    loc="upper center",
    facecolor="white",
    edgecolor="none"
)

# Add grid
ax.grid()
```

### ax.plot() - all step options

```python
fig, axs = plt.subplots(
    3, 1,
    sharex=True,
    sharey=True,
)
fig.suptitle(
    "ax.plot(x, y, drawstyle=...)",
    fontsize=20,
)

# Plot the points
for ax in axs:
    ax.scatter(
        range(len(y)),
        y,
        marker="X",
        s=275,
        linewidth=1.5,
        edgecolor="black",
        facecolor="white",
        zorder=100,  # crude way to keep the points always on top
    )

# Plot the step lines
axs[0].plot(
    y,
    color="red",
    drawstyle="steps-pre",
    label="ax.plot(..., drawstyle='steps-pre')"
)
axs[1].plot(
    y,
    color="blue",
    drawstyle="steps-mid",
    label="ax.plot(..., drawstyle='steps-mid')"
)
axs[2].plot(
    y,
    color="green",
    drawstyle="steps-post",
    label="ax.plot(..., drawstyle='steps-post')",
)

# Set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(-1, 6)
axs[2].set_xticks(range(0, 6))

# Add legend and grid
for ax in axs:
    ax.legend(loc="upper center", facecolor="white", edgecolor="none")
    ax.grid()
```

### ax.plot() & ax.fill_between()

```python
fig, axs = plt.subplots(
    3, 1,
    sharex=True,
    sharey=True
)
fig.suptitle(
    "ax.plot(...) & ax.fill_between(x, y, steps=...)",
)

# Plot the points
for ax in axs:
    ax.scatter(
        x, y,
        marker="X",
        s=275,
        linewidth=1.5,
        edgecolor="black",
        facecolor="white",
        zorder=100  # crude way to keep the points always on top
    )

# Plot the step lines and the fill
axs[0].plot(
    x, y,
    color="red",
    drawstyle="steps-pre",
)
axs[0].fill_between(
    x, y,
    step="pre",
    color="red",
    alpha=0.1,
    label="ax.fill_between(.., step='pre')"
)
axs[1].plot(
    x, y,
    color="blue",
    drawstyle="steps-mid",
)
axs[1].fill_between(
    x, y,
    step="mid",
    color="blue",
    alpha=0.1,
    label="ax.fill_between(.., step='mid')"
)
axs[2].plot(
    x, y,
    color="green",
    drawstyle="steps-post",
)
axs[2].fill_between(
    x, y,
    step="post",
    color="green",
    alpha=0.1,
    label="ax.fill_between(.., step='post')"
)

# Set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(0, 21)
axs[2].set_xticks(range(1, 21))

# Add legend and grid
for ax in axs:
    ax.legend(loc="upper center", facecolor="white", edgecolor="none")
    ax.grid()
```

### ax.stairs() - all data pre and post step

```python
fig, axs = plt.subplots(
    2, 1,
    sharex=True,
    sharey=True
)
fig.suptitle(
    "ax.stairs()",
)

# plot points
for ax in axs:
    ax.scatter(
        x, y,
        marker="X",
        s=275,
        linewidth=1.5,
        edgecolor="black",
        facecolor="white",
        zorder=100  # crude way to keep the points always on top
    )

axs[0].stairs(
    values=y,  # step height
    edges=[x[0], *x],  # edge width
    baseline=None,
    edgecolor="red",
    linewidth=2.5,
    label="steps-pre: duplicate 1st element in `edges=[x[0], *x]`",
)
axs[1].stairs(
    values=y,  # step height
    edges=[*x, x[-1]],  # edge
    baseline=None,
    color="green",
    linewidth=2.5,
    label="steps-post: duplicate last element in `edges=[*x, x[-1]]`",
)

# set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(0, 21)
axs[1].set_xticks(range(1, 21))

# Set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(0, 21)
axs[1].set_xticks(range(1, 21))

# Add legend and grid
for ax in axs:
    ax.legend(loc="upper center", facecolor="white", edgecolor="none")
    ax.grid()
```
