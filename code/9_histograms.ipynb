## Imports

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Mock Data

Let's generate made up distribution of heights and weights.

```python
sample_size = 2000

data_height = [
    random.normalvariate(mu=176,sigma=4)
    for _ in range(sample_size)
]
```

## 2. Plot: default histogram

```python
fig, ax = plt.subplots(
    figsize=(16,10),
    facecolor="white"
)
fig.suptitle(
    "Height"
)

ax.hist(
    data_height
)

# labels
ax.set_xlabel (
    "[cm]"
)
ax.set_ylabel(
    "count"
)

plt.tight_layout()

# plt.savefig(
#     fname="plot1.png",
#     format="png",
#     dpi=125
# );
```

## 3. Plot: tweak the appearance

* update colours (bars, spines, ticks and labels)
* update font size
* remove unnecessary spines

```python
grey_col = "#3d3846"  # GNOME dark 3
```

```python
fig, ax = plt.subplots(
    figsize=(16,10),
    facecolor="white"
)
fig.suptitle(
    "Height",
    fontsize=35,
    fontfamily="serif",
    weight="heavy",
    color=grey_col
)

ax.hist(
    data_height,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389"  # GNOME green 2
)

# labels
ax.set_xlabel (
    "[cm]",
    fontsize=20,
    color=grey_col
)
ax.set_ylabel(
    "count",
    fontsize=20,
    color=grey_col
)

# Update 'x' and 'y' axis tick labels size and colour
ax.tick_params(
    axis="both",
    labelsize=20,
    color=grey_col,  # tick colour
    labelcolor=grey_col
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

# Update spines colour
for position in ["bottom", "left"]:
    ax.spines[position].set_color(grey_col)

plt.tight_layout()

# plt.savefig(
#     fname="plot2.png",
#     format="png",
#     dpi=125
# );
```

## 5. Plot: update bin tick marks and add bar data labels

Let's use 12 bins as it's a reasonable number.

```python
fig, ax = plt.subplots(
    figsize=(16,10),
    facecolor="white"
)
fig.suptitle(
    "Height",
    fontsize=35,
    fontfamily="serif",
    weight="heavy",
    color=grey_col
)

n, bins, patches = ax.hist(
    data_height,
    bins=12,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389"  # GNOME green 2
)

# labels
ax.set_xlabel (
    "[cm]",
    fontsize=20,
    color=grey_col
)
ax.set_ylabel(
    "count",
    fontsize=20,
    color=grey_col
)
ax.bar_label(
    patches,
    fontsize=16,
    color="#26a269",  # GNOME green 5
)

# Set x-axis ticks
ax.set_xticks(bins)
ax.set_xticklabels([f"{value:.1f}" for value in bins])

# Update 'x' and 'y' axis tick labels size and colour
ax.tick_params(
    axis="both",
    labelsize=20,
    color=grey_col,  # tick colour
    labelcolor=grey_col
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

# Remove spines
for position in ["bottom", "left"]:
    ax.spines[position].set_color(grey_col)

plt.tight_layout()

# plt.savefig(
#     fname="plot4.png",
#     format="png",
#     dpi=125
# );
```

## 6. Plot: normalized data

* Data will be shown in relative terms

```python
fig, ax = plt.subplots(
    figsize=(16,10),
    facecolor="white"
)
fig.suptitle(
    "Height",
    fontsize=35,
    fontfamily="serif",
    weight="heavy",
    color=grey_col
)

n, bins, patches = ax.hist(
    data_height,
    bins=12,
    density=True,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389"  # GNOME green 2
)

# labels
ax.set_xlabel (
    "[cm]",
    fontsize=20,
    color=grey_col
)
ax.set_ylabel(
    "relative count",
    fontsize=20,
    color=grey_col
)
ax.bar_label(
    patches,
    fontsize=16,
    color="#26a269",  # GNOME green 5
    fmt="{:,.1%}"
)

# Set x-axis ticks
ax.set_xticks(bins)
ax.set_xticklabels([f"{value:.1f}" for value in bins])

# Update 'x' and 'y' axis tick labels size and colour
ax.tick_params(
    axis="both",
    labelsize=20,
    color=grey_col,  # tick colour
    labelcolor=grey_col
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

# Remove spines
for position in ["bottom", "left"]:
    ax.spines[position].set_color(grey_col)

plt.tight_layout()

# plt.savefig(
#     fname="plot5.png",
#     format="png",
#     dpi=125
# );
```

## 7. Plot: cummulative data

```python
fig, ax = plt.subplots(
    figsize=(16,10),
    facecolor="white"
)
fig.suptitle(
    "Height",
    fontsize=35,
    fontfamily="serif",
    weight="heavy",
    color=grey_col
)

n, bins, patches = ax.hist(
    data_height,
    bins=12,
    density=True,
    cumulative=True,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389"  # GNOME green 2
)

# labels
ax.set_xlabel (
    "[cm]",
    fontsize=20,
    color=grey_col
)
ax.set_ylabel(
    "relative cummulative count",
    fontsize=20,
    color=grey_col
)
ax.bar_label(
    patches,
    fontsize=16,
    color="#26a269",  # GNOME green 5
    fmt="{:,.1%}"
)

# Set x-axis ticks
ax.set_xticks(bins)
ax.set_xticklabels([f"{value:.1f}" for value in bins])

# Update 'x' and 'y' axis tick labels size and colour
ax.tick_params(
    axis="both",
    labelsize=20,
    color=grey_col,  # tick colour
    labelcolor=grey_col
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

# Remove spines
for position in ["bottom", "left"]:
    ax.spines[position].set_color(grey_col)

plt.tight_layout()

# plt.savefig(
#     fname="plot6.png",
#     format="png",
#     dpi=125
# );
```

## 8. Plot: two histograms

* add legend

### Data for the 2nd histogram

```python
data_height_2 = [
    random.normalvariate(mu=168,sigma=3.5)
    for _ in range(sample_size)
]
```

```python
fig, ax = plt.subplots(
    figsize=(16,10),
    facecolor="white"
)
fig.suptitle(
    "Height",
    fontsize=35,
    fontfamily="serif",
    weight="heavy",
    color=grey_col
)

ax.hist(
    data_height,
    bins=12,
    density=True,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389",  # GNOME green 2
    alpha=0.8,
    label="population 1"
)
ax.hist(
    data_height_2,
    bins=12,
    density=True,
    edgecolor="#613583",  # GNOME purple 5,
    facecolor ="#c061cb",  # GNOME purple 2
    alpha=0.8,
    label="population 2"
)

# labels
ax.set_xlabel (
    "[cm]",
    fontsize=20,
    color=grey_col
)
ax.set_ylabel(
    "relative count",
    fontsize=20,
    color=grey_col
)

# Update 'x' and 'y' axis tick labels size and colour
ax.tick_params(
    axis="both",
    labelsize=20,
    color=grey_col,  # tick colour
    labelcolor=grey_col
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

# Remove spines
for position in ["bottom", "left"]:
    ax.spines[position].set_color(grey_col)

ax.legend(
    fontsize=20,
    frameon=False
)

plt.tight_layout()

# plt.savefig(
#     fname="plot7.png",
#     format="png",
#     dpi=125
# );
```

## 9. Plot: two histograms with different style

```python
fig, ax = plt.subplots(
    figsize=(16,10),
    facecolor="white"
)
fig.suptitle(
    "Height",
    fontsize=35,
    fontfamily="serif",
    weight="heavy",
    color=grey_col
)

ax.hist(
    data_height,
    bins=12,
    density=True,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389",  # GNOME green 2
    alpha=0.65,
    label="population 1",
    histtype="stepfilled"
)
ax.hist(
    data_height_2,
    bins=12,
    density=True,
    edgecolor="#613583",  # GNOME purple 5,
    facecolor ="#c061cb",  # GNOME purple 2
    alpha=0.65,
    label="population 2",
    histtype="stepfilled"
)

# labels
ax.set_xlabel (
    "[cm]",
    fontsize=20,
    color=grey_col
)
ax.set_ylabel(
    "relative count",
    fontsize=20,
    color=grey_col
)

# Update 'x' and 'y' axis tick labels size and colour
ax.tick_params(
    axis="both",
    labelsize=20,
    color=grey_col,  # tick colour
    labelcolor=grey_col
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

# Remove spines
for position in ["bottom", "left"]:
    ax.spines[position].set_color(grey_col)

ax.legend(
    fontsize=20,
    frameon=False
)

plt.tight_layout()

# plt.savefig(
#     fname="plot8.png",
#     format="png",
#     dpi=125
# );
```

## Other histogram options

* horizontal
* log
* reduced bar width
* stacked
* variable bar width

```python
fig, axs = plt.subplots(
    2, 3,
    figsize=(16,10),
    facecolor="white"
)

axs.flat[0].set_title("Orientation:  horizontal")
axs.flat[0].hist(
    data_height,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389",  # GNOME green 2
    orientation="horizontal"
)

axs.flat[1].set_title("Log Y axis")
axs.flat[1].hist(
    data_height,
    bins=12,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389",  # GNOME green 2
    log=True
);

axs.flat[2].set_title("Reduced bar width")
axs.flat[2].hist(
    data_height,
    bins=12,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389",  # GNOME green 2
    rwidth=0.85
);

axs.flat[3].set_title("Stacked")
axs.flat[3].hist(
    [data_height, data_height_2],
    bins=np.arange(155, 195, 2.5),
    stacked=True,
    color=["#57e389", "#c061cb"]
)

axs.flat[4].set_title("Variable bar width")
axs.flat[4].hist(
    data_height,
    bins=[160, 170, 175, 177.5, 180, 195],
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389"  # GNOME green 2
)

axs.flat[5].set_axis_off()

plt.tight_layout()

# plt.savefig(
#     fname="plot9.png",
#     format="png",
#     dpi=125
# );
```
