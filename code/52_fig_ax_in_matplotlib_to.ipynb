## Plot elements (theory)

* `fig`: figure. It's like a frame that will hold our graphs.
* `axes`: canvas onto which we will plot our graph - they represent individual plots
  * `fig` can have many `axes`
  * `axes` always belong to one specific `fig`
  * `axes` shouldn't be confused with `axis`  (XY)

### `ax = fig.add_subplot()`

Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.add_subplot.html

Adds `axes` to a figure.

```python
fig = plt.figure()
ax = fig.add_subplot()
```

Most of the Matplotlib object accept *parameters* that will change the appearance of that object. We're going to feed two parameters to the `figure` object:

- `facecolor`: To better identify the background of the `figure` (not `axes`), we're adding a background color
  - The full pallette of *named colours* is available [here](https://matplotlib.org/stable/gallery/color/named_colors.html).
- `figsize`: to change the width and height of a plot

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)
ax = fig.add_subplot()
```

```python
type(ax)
```

By default, a single axis gets created and fills out the space of a figure (= fig.add_subplot(1, 1, 1)

To create a proper **subplot** in a grid, we need to feed some parameters to the function.

The parameters in the brackets denote:

* 1st digit: total nr of rows
* 2nd digit: total nr of columns
* 3rd digit: index (current axes)

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

# call signature nr 1:
ax_1 = fig.add_subplot(131)
ax_2 = fig.add_subplot(132)
ax_3 = fig.add_subplot(133)

# This signature allows to create up to 9 axes
```

To improve readability and also if we have more than 9 axes to adds, it's best to separate the numbers with `,`:

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

# call signature nr 2
ax_1 = fig.add_subplot(1, 3, 1)
ax_2 = fig.add_subplot(1, 3, 2)
ax_3 = fig.add_subplot(1, 3, 3)
```

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

ax_1 = fig.add_subplot(1, 3, 1)
ax_2 = fig.add_subplot(1, 3, 2)
ax_3 = fig.add_subplot(1, 3, 3)
```

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

ax_1 = fig.add_subplot(2, 3, 1)
ax_2 = fig.add_subplot(2, 3, 2)
ax_3 = fig.add_subplot(2, 3, 3)
ax_4 = fig.add_subplot(2, 3, 4)
ax_5 = fig.add_subplot(2, 3, 5)
ax_6 = fig.add_subplot(2, 3, 6)

ax_1.set_title('ax1')
ax_2.set_title('ax2')
ax_3.set_title('ax3')
ax_4.set_title('ax4')
ax_5.set_title('ax5')
ax_6.set_title('ax6')
```

You don't need to create `Axes` for every tile in the grid.

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

ax_1 = fig.add_subplot(2, 3, 1)
ax_3 = fig.add_subplot(2, 3, 3)
ax_5 = fig.add_subplot(2, 3, 5)

ax_1.set_title('ax1')
ax_3.set_title('ax3')
ax_5.set_title('ax5')
```

Adjacent `axes` can be merged in order to create more sofisticated layout. 
Indexes of `axes` to be merged need to be passed on as a `tuple`.

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

ax_1 = fig.add_subplot(2, 3, (1,2))  # tuple (first axes, last axes)
ax_3 = fig.add_subplot(2, 3, 3)
ax_4 = fig.add_subplot(2, 3, 4)
ax_5 = fig.add_subplot(2, 3, 5)
ax_6 = fig.add_subplot(2, 3, 6)

ax_1.set_title('ax1 and ax2 merged')
ax_3.set_title('ax3')
ax_4.set_title('ax4')
ax_5.set_title('ax5')
ax_6.set_title('ax6')
```

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

ax_1 = fig.add_subplot(2, 3, 1)
ax_2 = fig.add_subplot(2, 3, (2, 5))  # tuple (first axes, last axes)
ax_3 = fig.add_subplot(2, 3, 3)
ax_4 = fig.add_subplot(2, 3, 4)
ax_6 = fig.add_subplot(2, 3, 6)

ax_1.set_title('ax1')
ax_2.set_title('ax2 and ax5 merged')
ax_3.set_title('ax3')
ax_4.set_title('ax4')
ax_6.set_title('ax6')
```

Merging that spans multiple columns and rows:

* the `tuple` states the first and the last axes to encompass
* think of it as if you were drawing a rectangle starting at *first* position, and ending at *last* position

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

ax_1 = fig.add_subplot(2, 3, 1)
ax_2 = fig.add_subplot(2, 3, (2, 6))  # tuple (first axes, last axes)
ax_4 = fig.add_subplot(2, 3, 4)

ax_1.set_title('ax1')
ax_2.set_title('ax2, 3, 5, and 6 merged')
ax_4.set_title('ax4')
ax_6.set_title('ax6')
```

Weird inset grids are possible to create too:

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

ax_1 = fig.add_subplot(2, 3, 1)
ax_2 = fig.add_subplot(2, 3, (2, 6))  # tuple (first axes, last axes)
ax_4 = fig.add_subplot(2, 3, 4)
ax_6 = fig.add_subplot(2, 3, 6)

ax_1.set_title('ax1')
ax_2.set_title('ax2, 3, 5, and 6 merged')
ax_4.set_title('ax4')
ax_6.set_title('ax6')
```

### `ax = fig.add_axes([x, y, w, h])`

Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.add_axes.html



`fig.add_axes()` must be called with the size parameters, otherwise it won't create the `axes` and will through an error.

```python
# When one axes is added it seems to fill in the whole space regardless of the size
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine",
)

rect = 0, 0, 0.5, 0.5  #  x, y, width, heigh


ax = fig.add_axes(
    rect,
    facecolor="peachpuff"
)
```

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)

rect1 = 0, 0, 0.5, 0.5  #  x, y, width, heigh
ax1 = fig.add_axes(
    rect1,
    facecolor="peachpuff"
)

rect2 = 0.5, 0.5, 0.5, 0.5
ax2 = fig.add_axes(
    rect2,
    facecolor="lavender"
)
```

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine",
)

rect1 = 0, 0, 1, 1  #  x, y, width, heigh
ax1 = fig.add_axes(rect1, facecolor="peachpuff")
rect2 = 0.25, 0.25, 0.5, 0.5
ax2 = fig.add_axes(rect2, facecolor="lavender")
```

`zorder` parameter allows you to decide the order of axes in the figure. 

Elements with lower `zorder` are drawn first.

Order get's assigned as the `axes` get created:

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine",
)

rect1 = 0, 0, 0.6, 0.6  #  x, y, width, heigh
ax1 = fig.add_axes(
    rect1,
    facecolor="peachpuff"
)
rect2 = 0.4, 0.4, 0.6, 0.6
ax2 = fig.add_axes(
    rect2,
    facecolor="lavender"
)
```

Using `zorder` to change the order:

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine",
)

rect1 = 0, 0, 0.6, 0.6  #  x, y, width, heigh
ax1 = fig.add_axes(
    rect1,
    facecolor="peachpuff",
    zorder=2
)
rect2 = 0.4, 0.4, 0.6, 0.6
ax2 = fig.add_axes(
    rect2,
    facecolor="lavender",
    zorder=1
)
```

`fig.add_subplot()` and `fig.add_axes()` can be combined to create a subplot with and inset plot.

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)
ax = fig.add_subplot()

rect1 = 0.5, 0.5, 0.35, 0.3  #  x, y, width, heigh
ax1 = fig.add_axes(
    rect1,
    facecolor="peachpuff"
)
ax1.set_title("inset plot")
```

## Axes and its methods to create more Axes

### Inset: `ax = ax.inset_axes()`

Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.inset_axes.html

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)
ax = fig.add_subplot()

rect = 0.1, 0.1, 0.4, 0.3  #  x, y, width, heigh
ax_inset = ax.inset_axes(
    rect,
    facecolor="peachpuff"
)
ax_inset.set_title("inset plot");
```

Add limits and actual plot.

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)
ax = fig.add_subplot()
ax.plot([6, 3, 4, 3, 7, 3, 4, 4, 3, 2, 5, 2, 8, -1, 5])

rect = 0.1, 0.1, 0.4, 0.3  #  x, y, width, heigh
ax_inset = ax.inset_axes(
    rect,
    facecolor="peachpuff",
    xlim=(8, 12),
    ylim=(1, 6),
)
ax_inset.plot([6, 3, 4, 3, 7, 3, 4, 4, 3, 2, 5, 2, 8, -1, 5])

ax_inset.set_title("inset plot");
```

Finally, you can quickly add indicator of where the inset comes from.

Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.indicate_inset_zoom.html

```python
fig = plt.figure(figsize=(16, 8), facecolor="aquamarine")
ax = fig.add_subplot()
ax.plot([6, 3, 4, 3, 7, 3, 4, 4, 3, 2, 5, 2, 8, -1, 5])

rect = 0.1, 0.1, 0.4, 0.3  #  x, y, width, heigh
ax_inset = ax.inset_axes(
    rect,
    facecolor="peachpuff",
    xlim=(8, 12),
    ylim=(1, 6),
)
ax_inset.plot([6, 3, 4, 3, 7, 3, 4, 4, 3, 2, 5, 2, 8, -1, 5])
ax.indicate_inset_zoom(ax_inset)

ax_inset.set_title("inset plot");
```

### Twin plots: `ax2 = ax.twinx()` and `ax2 = ax.twiny()`

Docs: 
* https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.twinx.html
* https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.twiny.html

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)
ax = fig.add_subplot()
ax.plot([6, 3, 4, 3, 7, 3, 4, 4, 3, 2, 5, 2, 8, -1, 5], color="red")
ax.tick_params(
    axis="y",
    color="red",
    labelcolor="red",
)
ax_yy = ax.twinx()
ax_yy.plot([500, -100, 800, 200, 500, 200, 300, 400, 400, 300, 700, 300, 400, 300, 600], color="blue")
ax_yy.tick_params(
    axis="y", 
    color="blue",
    labelcolor="blue"
)
```

### Secondary axis: 

Docs:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.secondary_xaxis.html
* https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.secondary_yaxis.html

```python
def km2mile(x: int | float) -> float:
    """Convert km to miles."""
    return x * 0.62
```

```python
def mile2km(x: int | float) -> float:
    """Convert km to miles."""
    return x * 1.6
```

```python
fig = plt.figure(
    figsize=(16, 8),
    facecolor="aquamarine"
)
ax = fig.add_subplot()
ax.plot([6, 3, 4, 3, 7, 3, 4, 4, 3, 2, 5, 2, 8, -1, 5])
ax.set_xlabel("km")

sec_ax = ax.secondary_xaxis(
    location="top",
    functions=(
        km2mile, # function for converting units
        mile2km # inverse of the first function
    )
)
sec_ax.set_xlabel("miles");
```

### `fig, axs = plt.subplots_mosaic()`

Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot_mosaic.html

Use strings to create a gridded layout.

```python
layout = "abc"

fig, axs = plt.subplot_mosaic(
    mosaic=layout,
    figsize=(16, 8),
    facecolor="aquamarine",
)
```

```python
axs
```

```python
type(axs)
```

```python
axs["a"].set_title("a")
```

```python
fig
```

Build multi row and column grid expanding many rows and columns.

* `.` dot leaves the space blank

```python
layout = """
    abcc
    decc
    df.g
"""

fig, axs = plt.subplot_mosaic(
    mosaic=layout,
    figsize=(16, 8),
    facecolor="aquamarine",
)
```

Or you can use a list of lists for better labels.

```python
layout = [
    ["Top Left", "Center", "Top Right"],
    [".", "Center", "Bottom Right"],
]

fig, axs = plt.subplot_mosaic(
    mosaic=layout,
    figsize=(16, 8),
    facecolor="aquamarine",
)
```

You still have a possibility to pass on the parameter to the `Axe` object.

```python
layout = """
    ab
    cd
"""

fig, axs = plt.subplot_mosaic(
    mosaic=layout,
    figsize=(16, 8),
    facecolor="aquamarine",
    subplot_kw=dict(
        facecolor="peachpuff",
    ),
)
```

Or you can pass on dictionary of parameters so each `Axes` receives a unique set.

```python
layout = """
    ab
    cd
"""

fig, axs = plt.subplot_mosaic(
    mosaic=layout,
    figsize=(16, 8),
    facecolor="aquamarine",
    per_subplot_kw=dict(
        a=dict(facecolor="peachpuff"),
        b=dict(facecolor="cornflowerblue"),
        c=dict(facecolor="mintcream"),
        d=dict(facecolor="hotpink"),
    ),
)
```
