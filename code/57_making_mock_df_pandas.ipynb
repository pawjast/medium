# How to make mock up Data Frames

* First show the methods for creating DataFrames
* Then show random methods for generating data.

## Imports

```python
import pandas as pd
import numpy as np
```

## Example 1: single column from a list

```python
input_data = list("xyz")
input_data
```

```python
df = pd.DataFrame(
    input_data,
)
df
```

```python
df = pd.DataFrame(
    input_data,
    columns=["letters"]
)
df
```

## Example 2: single row from a list

Wrapping list into another list will switch `DataFrame` from being column oriented to being rows oriented.

```python
input_data = list("xyz")
input_data
```

```python
df = pd.DataFrame(
    [input_data],
)
df
```

```python
df = pd.DataFrame(
    [input_data],
    columns=["1st", "2nd", "3rd"]
)
df
```

## Example 3: list of lists

List of lists will always be rows oriented - each sublist makes a single row.

```python
input_data = [
    [1, 10, -1],  # Row 1
    [2, 20, -2],  # Row 2
    [3, 30, -3],  # Row 3
    [4, 40, -4],  # Row 4
]


input_data
```

```python
df = pd.DataFrame(
    data=input_data,
)
df
```

```python
df = pd.DataFrame(
    data=input_data,
    columns=["A", "B", "C"]
)
df
```

```python
df.T
```

```python
df.T.set_axis(list("ABCD"), axis=1).reset_index(drop=True)
```

## Example 7 - using `.from_dict()` methods

Summary:
* `column` orientation - `dict` keys become column names
* `index` orientation - `dict` keys become row index

### Column orientation (default)

```python
input_data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "London", "Paris"],
}
```

```python
df = pd.DataFrame.from_dict(
    data=input_data
)
df
```

### Row orientation

```python
# Existing data makes little sense in a column orientation
df = pd.DataFrame.from_dict(
    data=input_data,
    orient="index",
)
df
```

```python
#But timestamped data, e.g. log is better in row orientation
input_data = {
    "2024-01-01 10:00": ["INFO", "User logged in", "Alice"],
    "2024-01-01 10:05": ["WARNING", "Password attempt failed", "Bob"],
    "2024-01-01 10:10": ["ERROR", "Database timeout", "System"],
}
```

```python
df = pd.DataFrame.from_dict(
    data=input_data,
    orient="index",
    columns=["Level", "Message", "Actor"]
)
df
```

## Example 8: dict of dicts

When each entry in a dictionary is a full entity. E.g. when you pull data from APIs.

```python
input_data = {
    "user_1": {"Name": "Alice", "Age": 25, "City": "New York"},
    "user_2": {"Name": "Bob", "Age": 30, "City": "London"},
    "user_3": {"Name": "Charlie", "Age": 35, "City": "Paris"},
}
```

```python
# column orientation doesn't work for this data
df = pd.DataFrame.from_dict(
    data=input_data,
)
df
```

```python
df = pd.DataFrame.from_dict(
    data=input_data,
    orient="index"
)
df
```

```python
# If we want to keep the user ID as a stand alone column, we need to reset the index.
df = (
    pd.DataFrame
    .from_dict(
        data=input_data,
        orient="index"
    )
    .reset_index()
    .rename(columns={"index":"User"})
)

df
```
