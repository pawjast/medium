## Imports

````python
import matplotlib.pyplot as plt
````

## Data

*A mock up data.*

**Important:** `x` and `y` must be the same length!

````python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 1, -1, 1, -1, 1, 1]
````

_You've just mocked the data!_

__Well done you!__

## Plot

More info about the plotting function at 
[ax.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html)

````python
fig, ax  = plt.subplots(figsize=(8, 4))
ax.plot(x, y, lw=15, c="green")
````

To plot the data as the points use:

```
fig, ax  = plt.subplots()
ax.scatter(x, y)
```
